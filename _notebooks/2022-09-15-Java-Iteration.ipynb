```java
/*
 * Creator: Nighthawk Coding Society
 * Mini Lab Name: Hello Series,featuring Monkey Jumpers
 */

/**
 * Class for Monkeys: a 2D array of Monkeys
 * As well as method to print the Poem
 */
class AnimeLoop {
    //The area between class definition and the 1st method is where we keep data for object in Java
    String [][] animeArray;    //2D Array: AP CSA Unit 8: 2D array of strings
                            //2D array is like a grid [x][y]
                            // or like a spreadsheet [row][column]

    /**
     * Constructor initializes a 2D array of Monkeys
     */
    public AnimeLoop() {
        //Storing Data in 2D arrays
        animeArray = new String[][]{   //2D array above is just a name, "new" makes a container ("object")
                //Toji
                {
                "               ,@&@#",                
                "              @@@@@@@@@%",              
                "              @@@#@@&&@&",              
                "            .#@((/((/( /",              
                "        ,@@@@#/(##(  .",               
                "      /@@@@@@@@@@@@*",                  
                "    @@@@@@@@@@@@@@@@@@",                
                "  (@@@@@@@@@@@@@@@@@@@(",              
                "  #@@@@@@@@@@@@@@@@@@@@@",              
                "  @@@@@@@@@@@@@@@@@@@@@",               
                "  &@@@# @@@@@@@@@@@%%%#@",           
                "@@%%%%   @@@@@@@@@@%&&%#",             
                "  #&&&&%   %@@@@@@@@%&%%&#%",           
                "    &&&&. &@@@@@@@@@@%&&%%%",          
                "      *&#&&#%&.%%#%#@@#&&&&&",         
                "        %%%%#&&&/&&%%%%&* @@&&"   
                },
                //Megumi 1
                {
             "        @@@@@@@",                        
             "       @@@@@@@@@/&",                    
             "     %@@@@(@@@@@@@@",                    
             "       &@@@&@@@@@@@%",                   
             "    /@@**.,*.. * #@",                    
             "    @@&.,     , /.(",                    
             "   @@@@&@&, .*.@@, .,/",                 
             " ,@****#@@@*%(/@@@@/,  .",               
             "#@@@@@@@@&/(*/*%@@@%,,**,  *@",          
             "@@@@@@@@@@@*(#**(@@@@@, @@@@@@@",        
             " @@@@/*(&@@@*****@ @@@@. &@&@@@",        
             " ,/@@@@*@@#@@@@@%@      ,",              
             "  (#(%@@@@*@@@@@@@",                     
             "    (&@@@#%@@*@%**///((#   / ,      *",  
             "    *(@@@@@@@@**@@@@@*/@*@ ./*.*.   ,.", 
             "   ***@@@@@@@@@@@@@@@@@@@@       *",     
             "   /*@@@@@@@@@@@@&@@           #",       
             "  .**@@%&@%%@@@@@@@@             *",     
             "  *@@@@@@@@@&@@@@@@%              * (",  
             "   @@@@@@@@@@@@@@@@                , *" 
                },
                //Megumi 2
                {
          "          *  @",         
          "          @@*@@&@@(",      
          "        @@@@@@@@@@@@@@@%/",
          "         @@@@@@@@&@@@@@,", 
          "       /@@@@@@@/&@@@@@@#", 
          "         @@  ., ...*#@",   
          "         .  (      (/",    
          "    &@@@@/   ., .*/",     
          "   @@@@@@@@@  .@",         
          "   @@@@@&@@&@@#@@@@",      
          " .@&##%@%@@@@@@@@@",       
          "(&%%%%&@%&@@@@&@@@@",      
          "@%(##%@@@@@@@&&@@@@@",     
          "@%@@@@@@@&#%%@@@@@@@",     
          "@%@&@@@@@@@@%@@&%@@@&",    
          "&%%@@@@@@@@&&&%@@@&@@@",   
          "&&@##&@@@@@@@@&@@@&@@@@",  
          "@@&@@@@#@&@@@@@@@@@#@@@",  
          "@@&@@@&@%@@@@@@@@@@&%@@@", 
          "@@%%@@%@@%@@@@@@@%%@@&*(@",
          "  &@&@&%&@* ,&@@@@@&&@@(", 
          "   @@@&@*.     ,@@@@@@@@@",
          "    @@@@@@@@..    @@@@@@@",
          "      @@@@@@@@&#. ..@@@@"
                },
                //Megumi 3
                {
            "        .@@@@@@@@#*@",              
            "        .@&.@&@@@@@@@@@@",             
            "       %@@@@*@*@@@@,,@@@#",            
            "      #@@@@@@@@@@@@@@@@@@  (       ((",
            "       @@@,#.#@&&@&&@@@@@@&%(#(&((@%@",
            "         ,&., *,./#%.*#&&(&@@%/*@@&((",
            "         %&  . ,  *(/@@#@@@@@  @@ #((",
            "       ((@@%%,& &#/@@@@#(@@ *  *** .#",
            "     #&@#@#/. (  *,,(,%(,.//   , ,@@&",
            "   &%@@@&@@@@@@@@@@((@(/&   . /  ,  %",
            " #(#@(#((@(@@&@@@@@#@@@(",             
            "(((%@@#@(@%@@@@@@@@@@@&    /,   ,",   
            "(((%%@@@@(@%@@@@@@@@@@   (#@@@@@,",    
            "((@@@@@@@@@@@@@@@@&@&   @#@@@@@%&",   
            "  (@%%&&&@&@@@@%&@#&(  @@@@@@@@*",     
            "  .@&%#@@@@@@@@@%%#&(.(&#@@@@@(" 
                },
                //Megumi 4
                {
               "     (&&&&&&&&&  *",                               
               "     &&&&&&&&&&&&/  .",                              
               "     @@@&&#*%&&&@**",                                
               "       %,,.....,,",                                  
               "      .#*/....,.    ..",                             
               "     .&&&&&&%##(,..(*,,",                            
               "   *&&#&&&&&&&&*....",                               
               "&&&%&&@&&&&&&&&&...,#",                              
               "#&&&&&&&&&&&&&#&&&&&&#",                             
               " %&&%&&&&&&&&&&&&&&&&&&,",                           
               "  %&&&&&&%&&&&&&&@@&&&&&/",                          
               "   @&&&&&&&&&&@@@@@@@&&&@",                          
               "    @@&&&&&&@@@&&&&& @&&&",                          
               "     @@@&@@@&&&&&&&&",                               
               "     %@&&&&&&&&&&&&&&",                              
               "     &&&&&&&&&&&%&&&#",                              
               "    %&&&&&&&&&&&&&&&&",                              
               "    %&&&&&&&&&&&&%&&@&" 
                },

        };
    }

    public void printAnime() {
        //begin the poem
        System.out.println();
        System.out.println("Java Iteration Lab");

        // monkeys (non-primitive) defined in constructor knows its length
        int animeCount = animeArray.length;
        for (int i = animeCount; i >= 1; i--)  //loops through 2D array length backwards
        {

            //this print statement shows current count of Monkeys
            //  concatenation (+) of the loop variable and string to form a countdown message
            if (animeCount > 1) {
              System.out.println("There are " + (i-1) + " Megumi Fushiguros standing around...");
            } else {
              System.out.println("There is " + (i-1) + " Megumi Fushiguro standing around...");
            }
            

            //how many separate parts are there in a monkey monkey?
            for (int row = 1; row < animeCount; row++) {  //cycles through "cells" of 2d array

                /*cycles through columns to print
                each monkey part by part, will eventually print entire column*/
                for (int col = 0; col < animeArray[row].length; col++) {

                    // prints specific part of the monkey from the column
                    System.out.print(animeArray[row][col] + " ");

                    //this is new line between separate parts
                    System.out.println();
                }

                //this new line gives separation between stanza of poem
                System.out.println();
            }
            
            System.out.println("Here comes Toji Fushiguro who is a menace to society");

            for (int col = 0; col < animeArray[0].length; col++) {
              System.out.print(animeArray[0][col] + " ");
              System.out.println();
            }
            
            System.out.println();

            //countdown for poem, decrementing monkeyCount variable by 1
            animeCount -= 1;
        }

        //out of all the loops, prints finishing messages
        System.out.println("There are no more Megumi Fushiguros standing around");
        System.out.println("0000000000000000000000000000000000");
        System.out.println("             THE END              ");
    }

    /**
    * A Java Driver/Test method that is the entry point for execution
    */
    public static void main(String[] args)  {
        new AnimeLoop().printAnime();   //a new monkey list and output in one step
    }

}
AnimeLoop.main(null);
```
