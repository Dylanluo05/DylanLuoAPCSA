```java
import java.util.Scanner;

public class Level {
    int level;
    String levelStatus;
    public boolean goalReached() {
        if (levelStatus.equals("y")) {
            return true;
        } else {
            return false;
        }
    }
    public int getPoints() {
        if (level == 1 && goalReached() == true) {
            return 200;
        } else if (level == 2 && goalReached() == true) {
            return 100;
        } else if (level == 3 && goalReached() == true) {
            return 500;
        }
        return 0;
    }
}

public class Game {
    private Level levelOne = new Level();
    private Level levelTwo = new Level();
    private Level levelThree = new Level();

    /*public Game()
{  } */
    String bonusStatus;
    public void play() {
        System.out.println("Game Simulation");
        Scanner userInput = new Scanner(System.in);
        System.out.println("Did you complete level 1? (y/n)");
        levelOne.levelStatus = userInput.nextLine();
        levelOne.level = 1;
        System.out.println(levelOne.goalReached());
        System.out.println("Did you complete level 2? (y/n)");
        levelTwo.levelStatus = userInput.nextLine();
        levelTwo.level = 2;
        System.out.println(levelTwo.goalReached());
        System.out.println("Did you complete level 3? (y/n)");
        levelThree.levelStatus = userInput.nextLine();
        levelThree.level = 3;
        System.out.println(levelThree.goalReached());
        System.out.println("Is this a bonus game? (y/n)");
        bonusStatus = userInput.nextLine();
        //bonusStatus.isBonus();
    }
    
    public boolean isBonus() {
        if (bonusStatus.equals("y")) {
            return true;
        } else {
            return false;
        }
    }
    
    public int getScore() { 
        int score = 0;
        if (levelOne.goalReached() == true) {
            score += levelOne.getPoints();
            if (levelTwo.goalReached() == true) {
                score += levelTwo.getPoints();
                if (levelThree.goalReached() == true) {
                    score += levelThree.getPoints();
                }
            }
        }
        if (isBonus() == true) {
            score *= 3;
        }
        return score;
    } 

    //levelOne.goalReached();
    /*levelTwo.goalReached();
    levelThree.goalReached();*/

    /*public int playManyTimes(int num)
{ } */
    
    /*public static void main(String[] args) { 
        


        
        
        boolean levelTwoStatus = true;
        boolean levelThreeStatus = false;
        boolean bonus = false;

        int levelOnePoints = 1000;
        int levelTwoPoints = 1000;
        int levelThreePoints = 1000;

        int score = 0;
        if (levelOneStatus) {
            score += levelOnePoints;
            if (levelTwoStatus) {
                score += levelTwoPoints;
                if (levelThreeStatus) {
                    score += levelThreePoints;
                }
            }
        }

        if (bonus) {
            score *= 3;
        }

        System.out.println(score);
    }*/
}

public class testGame {
    public static void main(String[] args) {
        Game g = new Game();
        g.play();
        System.out.println(g.getScore());
    }
}
testGame.main(null);
```
